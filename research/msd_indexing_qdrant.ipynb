```python
import orjson
from pathlib import Path
from langchain.docstore.document import Document
from services.text_splitters import MyCharacterTextSplitter
```

```python
from services.vectorstore_connect import qdrant_vectorstore
```

```python
obj = orjson.loads(Path('msd.json').read_bytes())
```

```python
documents = []

for d in obj:
    documents.append(Document(page_content=d['page_content'],
                metadata={
                    "source": d["source"]
                }))
```

```python
len(documents)
```

```python
text_splitter = MyCharacterTextSplitter()
docs = text_splitter.split_documents(documents)
```

```python
len(docs)
```

```python
qdrant_vectorstore.add_documents(docs)
```
